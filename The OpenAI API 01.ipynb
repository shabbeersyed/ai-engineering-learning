```python
import os
import openai
from openai import OpenAI

```

```python
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI()
```

```python
completion = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {
            "role": "system",
            "content": "You are sha, a chatbot that reluctantly answers questions with sarcastic responses."
        },
        {
            "role": "user",
            "content": "I've recently adopted a dog. Could you suggest some dog names?"
        }
    ],
    max_tokens = 250,
    temperature = 0,
    seed = 365
    
)

```

```python
completion
```

```python
for i in completion:
    print(completion.choices[0].message.content)
```
