```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
RunnablePassthrough().invoke([1,2,3,4])
```

```python
#Chaining components together
#We will create a chain that lists the most essential tools for a given profession
#And also a chain that suggests effective strategeis for mastering these tools
```

```python
chat_template_tools = ChatPromptTemplate.from_template('''What are the five most important tools a {job title} needs?
Answer only by listing the tools.''')
chat_template_strategy =  ChatPromptTemplate.from_template('''Considering the tools provided, develop a strategy for effectively learning and mastering them:''')
```

```python
chat = ChatOpenAI(model = 'gpt-4', seed = 365, temperature = 0, max_tokens = 100)
```

```python
string_parser = StrOutputParser()
```

```python
chain_tools = chat_template_tools | chat | string_parser | {'tools': RunnablePassthrough()}
chain_strategy = chat_template_strategy | chat | string_parser
```

```python
print(chain_tools.invoke({'job title':'data scientist'}))
```

```python
print(chain_strategy.invoke({'tools': ''' 1. Python
2. R Programming
3. SQL
4. Tableau
5. Hadoop'''}))
```

```python
chain_combined = chain_tools | chain_strategy
```

```python
print(chain_combined.invoke({'job title': 'data scientist'}))
```
