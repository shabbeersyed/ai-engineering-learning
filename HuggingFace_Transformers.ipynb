# **Pre-trained Tokenizers**

```python
from transformers import AutoTokenizer
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I'm super excited to be learning about Large language models"
```

```python
input_ids = tokenizer(sentence)
```

```python
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
print(decoded_ids)
```

```python
tokenizer.decode(101)
```

```python
tokenizer.decode(102)
```

# **Huggingface and Pytorch/Tensorflow**

```python
from transformers import AutoTokenizer, AutoModelFor SequenceClassification
import torch

```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python

input_ids_pt = tokenizer(sentence, return_tensors="pt")
print(input_ids_pt)
```

```python

model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
with torch.no_grad():
    logits = model(**input_ids_pt).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```
