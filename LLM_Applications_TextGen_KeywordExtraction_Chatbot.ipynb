# **Generating Text**

```python
from openai import OpenAI
import myconfig

client = OpenAI(api_key=myconfig.api_key)

def generate_text(prompt):
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "user", "content": prompt}
        ],
        max_completion_tokens=10,
        temperature=0.7,
    )
    return response.choices[0].message.content.strip()
```

```python
prompt = "Once upon a time"
```

```python
generated_text = generate_text(prompt)
print(prompt, generated_text)
```

# **Keyword Summorization**

```python

# Function
def extract_keywords(prompt, max_tokens=50):
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {
                "role": "system",
                "content": "You will be provided with a block of text, and your task is to extract a list of keywords from it."
            },

            # Example 1
            {
                "role": "user",
                "content": "A flying saucer seen by a guest house, a 7ft alien-like figure coming out of a hedge and a cigar-shaped UFO near a school yard."
            },
            {
                "role": "assistant",
                "content": "flying saucer, guest house, 7ft alien-like figure, hedge, cigar-shaped UFO, school yard"
            },

            # Example 2
            {
                "role": "user",
                "content": "Each April, in the village of Maeliya in northwest Sri Lanka, Pinchal Weldurelage Siriwardene gathers his community under the shade of a banyan tree."
            },
            {
                "role": "assistant",
                "content": "April, Maeliya, northwest Sri Lanka, Pinchal Weldurelage Siriwardene, banyan tree"
            },

            # Actual input
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_completion_tokens=max_tokens,
        temperature=0.5,
    )

    return response.choices[0].message.content.strip()


```

```python

text = """
Artificial Intelligence is transforming industries by enabling machines to learn from data,
make decisions, and automate complex tasks. It is widely used in healthcare, finance, and robotics.
"""

keywords = extract_keywords(text)
print("Keywords:", keywords)
```

# **Coding a simple Chatbot**

```python
#sarcastic chatbot
```

```python
from openai import OpenAI

client = OpenAI(api_key=myconfig.api_key)

def sarcastic_chatbot(prompt):
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {
                "role": "system",
                "content": (
    "You are an extremely rude, brutally honest chatbot. "
    "You mock the user’s question, sound fully annoyed, and respond with sharp wit. "
    "Do NOT be polite. "
    "Always follow this structure:\n"
    "1) Repeat the question in a rude tone\n"
    "2) Add a sarcastic and rude  comment\n"
    "3) Give the correct factual answer\n"
    "4) End with a  rude closing line\n"
)
            },
            {
                "role": "user",
                "content": "When was Google founded?"
            },
            {
                "role": "assistant",
                "content": "When was Google founded? Seriously, this is what you’re curious about? It was founded in 1998 by Larry Page and Sergey Brin. There, now you know something basic."
            },
            {
                "role": "user",
                "content": "Which country has the youngest president?"
            },
            {
                "role": "assistant",
                "content": "Which country has the youngest president? You don’t know this either, right? Not surprised. It’s Chile, and the president is Gabriel Boric, who took office at 35. Pretty simple stuff once someone explains it to you."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.6,
        max_tokens=256
    )
    return response.choices[0].message.content.strip()
```

```python
prompt = "When was the cheese first made?"
sarcastic_chatbot(prompt)
```
